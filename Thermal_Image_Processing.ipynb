### Histogram equalization
#### Purpose: Enhances contrast in images by redistributing intensities across the histogram.
#### Function: Increases the global contrast of an image by stretching intensity range.

```python
# Apply histogram equalization to enhance image contrast
equalized_image = cv2.equalizeHist(thermal_image)

# Display the equalized image
plt.imshow(equalized_image, cmap='hot')
plt.title('Equalized Thermal Image')
plt.colorbar()
plt.axis('off')
plt.show()
```

### Thresholding
#### Purpose: Segments the image by dividing it into a binary image based on a threshold value.

#### Function: Converts grayscale images into binary images (black and white) based on intensity

```python
thresholded_image = cv2.inRange(thermal_image, 100, 200)

# Display the thresholded image
plt.imshow(thresholded_image, cmap='binary')
plt.title('Thresholded Thermal Image')
plt.axis('off')
plt.show()
```

### Median filtering
#### Purpose: Reduces noise while preserving edges.
#### Function: Replaces each pixel's value with the median value of its neighboring pixels

```python
median_filtered = cv2.medianBlur(thermal_image, 5)

# Display the median filtered image
plt.imshow(median_filtered, cmap='hot')
plt.title('Median Filtering')
plt.axis('off')
plt.show()
```

### Gaussian Blur
#### Purpose: Reduces noise and smoothens images by applying a Gaussian filter.
#### Function: Computes weighted average of neighboring pixels to blur edges and reduce high-frequency noise

```python
blurred_image = cv2.GaussianBlur(thermal_image, (5, 5), 0)

# Display the Gausian blur image
plt.imshow(blurred_image, cmap='hot')
plt.title('Gaussian Blur')
plt.axis('off')
plt.show()
```

### Erosion and Dilation (Morphological Operations):

#### Purpose: Modify shapes in the image like removing noise, isolating individual elements, or joining disjoint parts.
    

#### Erosion: Removes pixels from the edges.
    

```python
# Erosion
kernel = np.ones((5, 5), np.uint8)
erosion = cv2.erode(thermal_image, kernel, iterations=1)

# Display the erosion image
plt.imshow(erosion, cmap='hot')
plt.title('Erosion')
plt.axis('off')
plt.show()
```

#### Dilation: Adds pixels to the edges.

```python
# Dialation
dilation = cv2.dilate(thermal_image, kernel, iterations=1)

# Display the dialtion image
plt.imshow(dilation, cmap='hot')
plt.title('Dilation')
plt.axis('off')
plt.show()
```

### Colormap Converion

#### Purpose - translates pixel intensity values into colors, aiding visualization and interpretation of image data.

```python
colormap_image = cv2.applyColorMap(thermal_image, cv2.COLORMAP_JET)

# Display the colormap conversion image
plt.imshow(colormap_image)
plt.title('Colormap Conversion')
plt.colorbar()
plt.axis('off')
plt.show()
```
